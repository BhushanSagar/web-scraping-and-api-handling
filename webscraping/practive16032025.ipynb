```python
import requests 
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
```

```python
url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets'
```

```python
r= requests.get(url)
```

```python
r
```

```python
soup = bs(r.text,'lxml')
```

```python
soup 
```

```python
p_name=soup.find_all('a',class_ = 'title')
p_price=soup.find_all('h4',class_ = 'price float-end card-title pull-right')
p_description=soup.find_all('p',class_ = 'description card-text')
p_rating = soup.find_all('p', {'data-rating': True})
p_review=soup.find_all('p',class_ = 'review-count float-end')
```

```python
product_name=[]
product_price=[]
product_description=[]
product_rating=[]
product_review=[]
```

```python
for i in p_name:
    name= i.text
    product_name.append(name)
```

```python
for i in p_price:
    name= i.text
    product_price.append(name)
```

```python
for i in p_description:
    name= i.text
    product_description.append(name)
```

```python
for i in p_review:
    name= i.text
    product_review.append(name)
```

```python
for  i  in p_rating:
    product_rating.append(i['data-rating'])
 
```

```python
df = pd.DataFrame({
    "product name": product_name,
    "product price": product_price,
    "product description": product_description,
    "product rating": product_rating,
    "product review": product_review
})
```

```python
df
```

```python
df.index.name = "Product ID" 
```

```python
df.to_csv('product.csv')
```

```python
import requests 
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
import time
import random


```

```python
url = 'https://www.worldometers.info/coronavirus/countries-where-coronavirus-has-spread/'
```

```python
r= requests.get(url)
```

```python
r
```

```python
soup = bs(r.text,'lxml')
```

```python
table_heders = (soup.find_all('span',class_ = 'style1'))
```

```python
table_header = []
```

```python
for i in table_heders:
    name = i.text
    table_header.append(name)
```

```python
# create dataframe
```

```python
df=pd.DataFrame(columns=table_header)
```

```python
table= soup.find_all('tr')
```

```python
for i in table[1:]:
    row=i.find_all('td')
    e_row=[t.text for t in row]
    lens=len(df)
    df.loc[lens]=e_row
```

```python
df
```

```python
df.index.name = 'id'
```

```python
df
```

```python
df.to_csv('covid 19 spread.csv')
```
