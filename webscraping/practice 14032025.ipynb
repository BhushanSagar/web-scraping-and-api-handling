```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets'
```

```python
r= requests.get(url)
```

```python
r
```

```python
datas=bs(r.text,'lxml')
```

```python
p_price = datas.find_all('h4',class_= 'price float-end card-title pull-right')
p_name = datas.find_all('a',class_= 'title')
p_description = datas.find_all('p',class_= 'description card-text')
p_review = datas.find_all('p',class_= 'review-count float-end')
p_rating = datas.find_all('p',class_= 'data-rating')
```

```python
p_price
```

```python
product_price = []
product_name = []
product_description = []
product_review = []
for i in p_price:
    name = i.text
    product_price.append(name)



for j in p_name:
    name = j.text
    product_name.append(name)


for k in p_description:
    name = k.text
    product_description.append(name)


for l in p_review:
    name = l.text
    product_review.append(name)

```

```python
print(len(product_name))
print(len(product_price))
print(len(product_description))
print(len(product_review))
```

```python
#create dataframe


df = pd.DataFrame({
    "product name": product_name,
    "product price": product_price,
    "product description": product_description,
    "product review": product_review
})
```

```python
df
```

```python
df.to_csv('product_data.csv', index=False)
```
