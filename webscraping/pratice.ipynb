```python
import requests

from bs4 import BeautifulSoup
```

```python
url = "https://webscraper.io/test-sites/e-commerce/allinone"
```

```python
r= requests.get(url)
```

```python
print(r)
```

```python
soup = BeautifulSoup(r.text,"lxml")
```

```python
print(soup)
```

```python
soup_data=soup.find("h4",{"class":"price float-end card-title pull-right"})
```

```python
print(soup_data)
```

```python
#now wth find_all


soup_data_all=soup.find_all("h4",{"class":"price float-end card-title pull-right"})
```

```python
print(soup_data_all)
```

```python
for i in soup_data_all:
    print(i)
```

```python
# to data from multiple tags
```

```python
print(soup.find_all(["h4","a"]))
```

```python
# to get all ids
print(soup.find_all(id=True))
```

```python
# find data with similar

print(soup.find_all(string="Screenshots"))
```

```python
# find all with regex

import re 
```

```python
print(soup.find_all(string=re.compile("commerce")))
```

```python
#to find multiple pattern
```

```python
print(soup.find_all(string=["commerce","lenovo"]))
```

```python
# to take desciption
```

```python
print(soup.find_all(class_ = re.compile("description")))
```

```python
#to get specific data
```

```python
print(soup.find_all("p",class_ = re.compile("description")))
```

```python
print(soup.find_all("p",class_ = re.compile("description"),limit=2))
```
