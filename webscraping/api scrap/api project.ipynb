```python
import requests
import pandas as pd

def get_data_and_save():
    url = "https://api.freeapi.app/api/v1/public/quotes"
    response = requests.get(url)

    if response.status_code == 200:
        data_json = response.json()
        print(data_json)  # Debugging: Print response JSON
        df = pd.json_normalize(data_json, 'data')
        return df
      




```

```python
# Call function
get_data_and_save()
```
